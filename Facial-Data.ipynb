```python
from ultralytics import YOLO
# Load the YOLO model
model = YOLO('yolov8n.pt')

results = model.train(
    data='/kaggle/input/affectnet-yolo-format/YOLO_format/data.yaml',  # Path to dataset YAML
    epochs=50,                         # Number of epochs
    imgsz=96,                         # Image size
    batch=16,                          # Batch size
    device=0                           # Use GPU if available
)
```

```python
import shutil

# Source folder and target zip file
source_folder = "/kaggle/working/runs"
output_zip = "/kaggle/working/runs.zip"

# Create zip file
shutil.make_archive(base_name=output_zip.replace(".zip", ""), format='zip', root_dir=source_folder)

print("Zipping complete:", output_zip)
```
